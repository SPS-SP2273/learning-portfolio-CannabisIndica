# A simple plot

## Let's look at some code

````python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
````

Here is how the different functions work.

### ```plt.plot```

This function plots a graph where points are joined by a line. You can change this to something else like ```plt.scatter``` to change the type of graph. This command defines what kind of plot will be plotted. 

````python
plt.scatter(x,y_1)
````

Now focusing on arguments inside ```plt.plot```. **ALL the arguments inside this function are optional**. The first argument is the variable that will be plotted on the x-axis and the second argument is the variable that will be plotted on the y-axis. The ```label``` is the legend of the graph, the ```color``` or ```c``` is the color of the line being drawn, the ```linestyle``` or ```ls``` is the type of line being drawn, and the ```linewidth``` or ```lw``` is the thickness of the line being drawn.  

````python
plt.plot(x, y_1, label='random trial',
         c='cyan', ls='solid',lw=2)
````

### ```plt.legend```

As you can see in the figure above, you cannot see the legend anywhere. To make sure the legend appears, you need to add the ```plt.legend``` command.

````python
plt.plot(x, y_1, label='random trial',
         c='cyan', ls='solid',lw=2)
plt.legend()
````

As you can see the argument inside the function is **optional**. You can change the location using ``loc``, or you can also manually set the location (which is tiring but useful in some complicated plots).

````python
plt.plot(x, y_1, label='random trial',
         c='cyan', ls='solid',lw=2)
plt.legend(loc='lower right')
````

### ```plt.xlabel``` and ```plt.ylabel```

These commands can be used to put axis-labels. The font size can also be changed. 

````python
plt.plot(x, y_1, label='random trial',
         c='green', ls=':',lw=2)
plt.legend()
plt.xlabel('cool x-label')
plt.ylabel('cooler y-label')
````

### ```plt.title```

As the name suggests, plots the best title possible. You can set its location as well. 

````python
plt.plot(x, y_1, label='random trial',
         c='green', ls=':',lw=2)
plt.legend()
plt.xlabel('cool x-label')
plt.ylabel('cooler y-label')
plt.title('Most complicated graph')
````

### ```plt.grid```

This creates a grid in the graph. 

````python
plt.plot(x, y_1, label='random trial',
         c='green', ls=':',lw=2)
plt.legend()
plt.xlabel('cool x-label')
plt.ylabel('cooler y-label')
plt.title('Most complicated graph')
plt.grid()
````

However this grid is so dark, so we can cause its transperancy to increase using the ```alpha``` argument. 

````python
plt.plot(x, y_1, label='random trial',
         c='green', ls=':',lw=2)
plt.legend()
plt.xlabel('cool x-label')
plt.ylabel('cooler y-label')
plt.title('Most complicated graph')
plt.grid(alpha=.5)
````

### ```plt.show```

If you have seen usual matplotlib code, they always include ```plt.show()``` at the end of the code. However, **this is not required in Jupyter, but should still add to use in other interfaces**. 

## Adding another plot

````python
plt.plot(x, y_1, label='random trial',
         c='green', ls=':',lw=2)
plt.plot(x,y_2, label = 'another trial', c='red',ls='--',lw=3)
plt.legend()
plt.xlabel('cool x-label')
plt.ylabel('cooler y-label')
plt.title('Most complicated graph')
plt.grid(alpha=.5)
````

## Yet another plot but with error bars

All good science diagrams will include error bars. To include the most BASIC error bars we can use ```plt.errorbar```. There are more optional arguments, but we shall cover them in the future if i feel like it. 

````python
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, label='random trial',
         c='green', ls=':',lw=2)
plt.plot(x,y_2, label = 'another trial', c='red',ls='--',lw=3)
plt.errorbar(x,y_3,xerr=x_error,yerr=y_error,label='error stuff',c='blue')
plt.legend()
plt.xlabel('cool x-label')
plt.ylabel('cooler y-label')
plt.title('Most complicated graph')
plt.grid(alpha=.5)
````

# Better with NumPy

This is where NumPy shines. As said in the website we will convert everything to arrays. 

````python
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, c='red', ls=':', label='cool value')
plt.plot(np_x, np_y_2, c='blue', label='uncool value')
plt.grid(alpha=.75)
plt.legend(loc='best')
````

## Adding mathematical functions

Because we have numpy arrays now, we can use mathematical functions on the data sets and plot cool graphs. 

````python
cosine_graph = np.cos(np_x)
x3 = np_x**3

plt.plot(x,cosine_graph, label='cosine')
plt.plot(x,x3,label=f'x-cubed')
plt.legend()
plt.grid(alpha=0.25)
````

## We need another axis!

First of all, different graphs on save axis will be bad, so we can have two different y-axis using ```plt.twinx```.

````python
cosine_graph = np.cos(np_x)
x3 = np_x**3

plt.plot(x,cosine_graph, label='cosine',c='red')
plt.legend(loc='lower right')
plt.ylabel('cosine')
plt.twinx()
plt.plot(x,x3,label='x-cubed')
plt.legend()
plt.ylabel('x-cubed')
````

As seen above, this looks VERY UGLY, and this is because **our x-axis is a set of discrete variables and plt.plot is just joining the points using a straight line**. Therefore we will use ```np.linspace``` which linearly spaces out the given number of points between start and end. 

````python
x = np.linspace(0,4*np.pi,1000)
cosine_graph = np.cos(x)
x3 = x**3

plt.plot(x,cosine_graph, label='cosine',c='red')
plt.legend(loc='lower right',fontsize=9)
plt.ylabel('cosine')
plt.twinx()
plt.plot(x,x3,label='x-cubed')
plt.legend(loc='lower left',fontsize=9)
plt.ylabel('x-cubed')
````

Honestly, its really irritating to plot this way. There are much better options.

# Saving to disc

To save a figure, we can use ```plt.savefig``` to save it. The first argument is the path and the keyword argument is ```dpi``` which is dots per inch. The higher the dpi, the greater the resolution (and consequently the size) of your figure. If instead of path only name is given, it will get saved on the current working directory. We can specify the save location as well such as ```S://Academics/funnyplot.png```.

````python
x = np.linspace(0,4*np.pi,1000)
cosine_graph = np.cos(x)
x3 = x**3

plt.plot(x,cosine_graph, label='cosine',c='red')
plt.legend(loc='lower right',fontsize=9)
plt.ylabel('cosine')
plt.twinx()
plt.plot(x,x3,label='x-cubed')
plt.legend(loc='lower left',fontsize=9)
plt.ylabel('x-cubed')
plt.savefig('ugly-plot.pdf',dpi = 1000)
````

Also note that the file is OVERWRITTEN everytime you re-run the code.

## Plotting data from files

We use the ```np.loadtxt``` to load txt files and skip the rows that contains things we don't want. Then we use ```plt.ylim``` to set the minimum and maximum value of y-axis.

````python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-4, 4])
````

We can use the ```plt.hlines``` function to create a dashed horizontal line for our data. 

````python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-4, 4])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
````

## Styles

We can change styles. There are a lot, see below. 

````python
plt.style.available
````

My personal favourite is still classic idk why (don't come at me please).

````python
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.style.use('classic')
plt.plot(date, anomaly, alpha=.25)
plt.ylim(-3.5, 3.5)
plt.xlim(1875,2030)
plt.ylabel('Temperature Anomaly')
plt.xlabel('Date')
plt.title('Temperature Anomaly \n (Relative to average from Jan 1951 - Dec 1980)')
plt.hlines(0, date[0], date[-1], linestyle='--', colors='black')
````
